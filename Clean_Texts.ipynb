## Iterate through all files.

```python
directories = [x[0] for x in os.walk(mishnah_path)]
```

```python
english = [i for i in directories if 'English' in i and not 'Commentary' in i] 
hebrew = [i for i in directories if 'Hebrew' in i and not 'Commentary' in i] 
```

```python
for dir in english:
    mishnah = open(dir + '/merged.txt', 'rb').read().decode(encoding='utf-8')
    mishnah = mishnah[mishnah.find('\n\n\n\n'):][2:]
    mishnah = re.sub(r'\n\nChapter .*\n', '\n', mishnah)[2:]
    #mishnah = re.sub('\n\n', '\n', mishnah)
    mishnah += '\n\n'
    
    text_file = open(dir + '/clean.txt', 'wt')
    n = text_file.write(mishnah)
    text_file.close()
```

```python
for dir in hebrew:
    mishnah = open(dir + '/merged.txt', 'rb').read().decode(encoding='utf-8')
    mishnah = mishnah[mishnah.find('\n\n\n'):][2:]
    mishnah = re.sub(r'\n\nChapter .*\n', '\n', mishnah)[2:]
    #mishnah = re.sub('\n\n', '\n', chapterless_mishnah)
    mishnah = re.sub('\n\n', '\n', mishnah)[:-2]
    mishnah += '\n\n'
    
    text_file = open(dir + '/clean.txt', 'wt')
    n = text_file.write(mishnah)
    text_file.close()
```
